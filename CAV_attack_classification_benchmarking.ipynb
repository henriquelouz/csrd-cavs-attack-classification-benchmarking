### Explore Dataset

```python
import pandas as pd
```

```python
columnames = ["timestamp", "can_id", "dlc", "data_0", "data_1", "data_2", "data_3", "data_4", "data_5", "data_6", "data_7", "flag"]
```

```python
dos_df = pd.read_csv("DoS_dataset.csv", names=columnames)
dos_df.head(5)
```

```python
fuzzy_df = pd.read_csv("Fuzzy_dataset.csv", names=columnames)
fuzzy_df.head(5)
```

```python
gear_df = pd.read_csv("gear_dataset.csv", names=columnames)
gear_df.head(5)
```

```python
rpm_df = pd.read_csv("RPM_dataset.csv", names=columnames)
rpm_df.head(5)
```

```python
df = pd.concat([dos_df, fuzzy_df, gear_df, rpm_df])
df.shape
```

```python
df["flag"].unique()
```

```python
df = df.dropna()
print(df["flag"].unique())
print(df.shape)
```

```python
ids = rpm_df["0316"]
rpm_df[ids.isin(ids[ids.duplicated()])].sort_values("0316")
```

```python
df['can_id'] = df['can_id'].apply(lambda x: int(x, base=16))
df['data_0'] = df['data_0'].apply(lambda x: int(x, base=16))
df['data_1'] = df['data_1'].apply(lambda x: int(x, base=16))
df['data_2'] = df['data_2'].apply(lambda x: int(x, base=16))
df['data_3'] = df['data_3'].apply(lambda x: int(x, base=16))
df['data_4'] = df['data_4'].apply(lambda x: int(x, base=16))
df['data_5'] = df['data_5'].apply(lambda x: int(x, base=16))
df['data_6'] = df['data_6'].apply(lambda x: int(x, base=16))
df['data_7'] = df['data_7'].apply(lambda x: int(x, base=16))
```

```python
df.head(5)
```

```python
df.dtypes
```

## SVM

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['data_0', 'data_1', 'data_2', 'data_3', 'data_4', 'data_5', 'data_6', 'data_7']], df['flag'], test_size=0.3, random_state=109) # 70% training and 30% test
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
